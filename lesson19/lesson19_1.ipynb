```python
import numpy as np
import pandas as pd
scores = np.random.randint(50,101,size=(50,5))
scores_pd = pd.DataFrame(scores,
             columns=['國文','英文','數學','地理','歷史'],
             index = range(1,51)
            )
scores_pd
```

```python
scores_pd.index.name='學號'
scores_pd.columns.name='科目'
scores_pd
```

```python
#axis：軸
#axis=0：由上而下開始計算
#axis=1：由左而右開始計算
```

```python
#統計的實體方法：sum()
scores_pd['總分'] = scores_pd[['國文','英文','數學','地理','歷史']].sum(axis=1)
scores_pd
scores_pd['平均'] = scores_pd[['國文','英文','數學','地理','歷史']].mean(axis=1) #mean：平均
scores_pd
```

```python
scores_pd['排名'] = scores_pd['總分'].rank(method='min',ascending=False)
scores_pd
```

```python
scores_pd.sort_values(by='排名') #排序 buy='欄位名稱'
```

```python
def formate60(v:int): #V：變數，此設定為分數
    if v >= 60:
        return "及格"
    else:
        return "不及格"

scores_pd[['國文','英文','數學','地理','歷史']].style.format(formate60)
```
