```python
import pandas as pd

scores = [
    {'chinese':78,'english':80,'math':92,'history':75,'discover':85},
    {'chinese':92,'english':60,'math':85,'history':62,'discover':73}
]
dataFrame = pd.DataFrame(index=["rober",'jenny'],data = scores)
display(dataFrame.T)
```

```python
display(dataFrame.T.plot())
display(dataFrame.plot())
```

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font = FontProperties(fname=r'c:\WINDOWS\Fonts\GENSENROUNDED-R.TTC',size=24)
figure = plt.figure()  #畫布
axes = figure.add_subplot(1,1,1) #傳出圖表
dataFrame.T.plot(ax=axes,style=['-','--'],marker = 'o')  #畫在這個畫布上
axes.set_title("分數比較表",fontproperties=font)
axes.grid(visible=True,axis='y') #grid  #axie可以自己來管，用dataFrame.plot來畫
plt.show()

#自動加入axes.legend
```
