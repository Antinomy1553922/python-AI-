```python
import pandas as pd

scores = [
    {'chinese':78,'english':80,'math':92,'history':75,'discover':85},
    {'chinese':92,'english':60,'math':85,'history':62,'discover':73}
]
dataFrame = pd.DataFrame(index=["rober",'jenny'],data = scores)
display(dataFrame.T)
```

```python
display(dataFrame.T.plot())
display(dataFrame.plot())
```

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font = FontProperties(fname=r'c:\WINDOWS\Fonts\GENSENROUNDED-R.TTC',size=24)
figure = plt.figure()  #
axes = figure.add_subplot(1,1,1) #
dataFrame.T.plot(ax=axes,style=['-','--'],marker = 'o')  #
axes.set_title("",fontproperties=font)
axes.grid(visible=True,axis='y') #grid  #axiedataFrame.plot
plt.show()

#axes.legend
```
