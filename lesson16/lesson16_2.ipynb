```python
import matplotlib.pyplot as plt
#建立一個figure的實體並傳送給我們
#figure就是一個「畫布」

figure = plt.figure()  #接收figure
axes = figure.add_subplot(1,1,1)
x = [0,1,2]
y = [3,4,5]
axes.plot(x,y,'--bo') #axes的實體方法
plt.show
```

```python
import matplotlib.pyplot as plt

figure = plt.figure()
axes1 = figure.add_subplot(1,3,1)
axes1.set_title("title1")
axes2 = figure.add_subplot(1,3,2)
axes2.set_title("title2")
axes3 = figure.add_subplot(1,3,3)
axes3.set_title("title3")
```

```python
#一次建立兩個畫布

figure1 = plt.figure()
axes1 = figure1.add_subplot(1,1,1) #figure1裡面的圖表
x1 = [0,1,2]
y1 = [3,4,5]
axes1.plot(x1,y1)

figure2 = plt.figure()  #建立第二個figure
axes2 = figure2.add_subplot(1,1,1)  #第二個圖表
x2 = [2,3,6]
y2 = [4,3,7]
axes2.plot(x2,y2)

plt.show
```
