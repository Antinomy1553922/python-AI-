```python
import numpy as np
import pandas as pd

df = pd.DataFrame({'key1':['a','a','b','b','a'],
                    'key2':['one','two','one','two','one'],
                    'data1':np.random.randn(5),
                    'data2':np.random.randn(5)})
df
```

```python
df1 = df.set_index(keys=['key1','key2'])
df1
```

```python
df1.columns.name = 'data'
df1
```

```python
df2 = df1.groupby(level=['key1','key2']).mean()
df2
#groupby會自動把欄位變索引，不需要set_index
```

```python
df2.unstack(level='key1')
```
